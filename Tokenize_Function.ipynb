```python
from transformers import AutoTokenizer
from datasets import load_dataset

# Load a dataset for demonstration purposes
dataset = load_dataset("imdb")

# Instantiate a tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Define the tokenize function
def tokenize_fn(batch):
    # Access the 'text' key from the batch dictionary
    sentences = batch['text']

    # Tokenize the sentences using the tokenizer
    tokenized_sentences = tokenizer(sentences, truncation=True)

    # Print the tokenized output for the first example in the batch
    print("Original sentence:", sentences[0])
    print("Tokenized sentence:", tokenized_sentences['input_ids'][0])

    return tokenized_sentences

# Tokenize the dataset using the map function
tokenized_dataset = dataset.map(tokenize_fn, batched=True)
```
